```python
fig = go.Figure()

fig.add_trace(
        go.Box(y=df["Pregnancies"])
)

fig.show()
```

```python
fig= make_subplots(rows=3, cols=3)

cols_i=0
for i in range(1,4):
    for j in range(1,4):
        fig.add_trace(
            go.Box(
                y=df[cols[cols_i]],
                name=cols[cols_i]
            ),
            row=i, col=j
        )
        cols_i+=1

fig.update_layout({'height': 900})
fig.show()
```

# Visualization

```python
#normalize the info about each Outcome to compare
df_norm=pd.DataFrame()

for col in df.columns:
    df_norm[col]=df[col]/df[col].max()
    
df_0=df_norm[df_norm.Outcome==0]
df_1=df_norm[df_norm.Outcome==1]
```

```python
features=list(cols)
features.pop()

fig= go.Figure()

fig.add_trace(go.Scatterpolar(
    r=list(df_0.mean()),
    theta=features,
    name='0'
))

fig.add_trace(go.Scatterpolar(
    r=list(df_1.mean()),
    theta=features,
    name='1'
))

fig.show()
```

The outcomes True have higher values on alls features, particularly on Glucose and Pregnancies
